```python
import numpy as np
import pandas as pd
import re
import spacy
import gensim.downloader as api

wv = api.load("word2vec-google-news-300")
df = pd.read_csv("temp.csv")
nlp = spacy.load("en_core_web_lg")

df.head()
```

```python
def preprocess(array):
    matches = re.findall(r"[a-zA-Z]*",array)
    res = []
    for val in matches:
        if val == '':
            continue
        else:
            res.append(val)
            
    s = ' '.join(res).lower().strip()
    return s

def lemmatize(s):
    filtered = []
    doc = nlp(s)
    for token in doc:
        if token.is_punct or token.is_stop:
            continue
        else:
            filtered.append(token.lemma_)
    
    return wv.get_mean_vector(filtered)


# def mergeall(array1, array2, array3):
#     s = array1 + array2 + array3
#     return s.lower().strip()

# def low(array):
#     return array.lower()

# s = preprocess(df['genres'][0])
# df['new'][0] = df['author'][0] + " " + df['description'][0] + " " + s
# print(df['new'][0])
# print()
# dfk['description'] = dfk['description'].apply(low)
# dfk.head()
```
